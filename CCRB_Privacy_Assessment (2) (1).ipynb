```python
def gender_bin(gender_col_vals):
    if gender_col_vals in ["Male", "Female"]:
        val = "M/F"
    elif gender_col_vals == 'TGNC / Other':
        val = 'TGNC/Other'
    else:
        val = "null"
    return val
```

```python
def race_bin(race_col_vals):
    if race_col_vals in ['Black', 'White', 'Hispanic', 'Asian', 'American Indian']:
        val = "Reported"
    elif race_col_vals == "Other":
        val = "Other"
    elif race_col_vals == "Refused":
        val = "Refused"
    elif race_col_vals == "Unknown":
        val = "Unknown"
    else:
        val = "null"
    return val
```

```python


def show_values(axs, orient="v", space=.01):
    def _single(ax):
        if orient == "v":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() / 2
                _y = p.get_y() + p.get_height() + (p.get_height()*0.01)
                value = '{:.2f}'.format(p.get_height()).lstrip('0.00')
                ax.text(_x, _y, value, backgroundcolor='white', fontsize=10, animated=True, ha="center", va="center_baseline") 
        elif orient == "h":
            for p in ax.patches:
                _x = p.get_x() + p.get_width() + float(space)
                _y = p.get_y() + p.get_height() - (p.get_height()*0.5)
                value = '{:.2%}'.format(p.get_width())
                ax.text(_x, _y, value, ha="left")

    if isinstance(axs, np.ndarray):
        for idx, ax in np.ndenumerate(axs):
            _single(ax)
    else:
        _single(axs)
```

```python
import matplotlib.pyplot as plt

# depict illustration
plt.figure(figsize=(8,8))
colors_list = ['Red','Orange', 'Pink']
plt.grid(color='gray', linewidth=1, axis='both', alpha=0.5)
graph = plt.bar(data.Column_Name,data.Percentage, color = colors_list)
plt.title('Percentage Populated with Demographic Information')
    
ax = plt.gca()
ax.set_ylim([0, 100])
ax.set_ylabel('Percent Populated')
#ax.set_xlabel('Field Names')
#ax.set_xticks(data.Column_Name)
#ax.set_xticklabels(data.Column_Name)
#ax.set_yticks(data.Percentage)

show_values(ax, space=0)

plt.show()
 
```
